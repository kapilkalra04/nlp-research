```python
def read_input(input_file):
    # removing '__label__n' from the start of each line
    CUSTOM_FILTERS = [lambda x: x[10:].lower(), strip_tags, strip_punctuation, strip_multiple_whitespaces, strip_numeric]
    with io.open (input_file, encoding="utf8", errors="ignore") as f:
        for i, line in enumerate (f):
            if (i%10000==0):
                print i
            # do some pre-processing and return a list of words for each review text
            yield preprocess_string(line, CUSTOM_FILTERS)
        print i
```

```python
documents1 = list(read_input(input_1))
```

```python
documents2 = list(read_input(input_2))
```

```python
model = gensim.models.Word2Vec (documents1, size=200, window=10, min_count=2, workers=8)
model.train(documents1,total_examples=len(documents1),epochs=20)
```
